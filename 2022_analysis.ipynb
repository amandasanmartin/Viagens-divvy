```python
#Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
cores_vivas = sns.color_palette("bright")
```

Porcentagem por tipo de membros

```python
#Caminho do arquivo
cam = '/content/drive/MyDrive/Analises/2022/tipo_membro.csv'
dfm = pd.read_csv(cam)
```

```python
# Agrupamento
labels = dfm['tipo_uso']
sizes = dfm['total']

#Conversão para porcentagem
def func_pct_abs(pct, all_vals):
    absolute = pct / 100. * sum(all_vals)
    return f"{pct:.1f}%\n({absolute/1e6:.1f}M)"

#Desenho do gráfico
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=None,
    autopct=lambda pct: func_pct_abs(pct, sizes),
    startangle=90,
    colors=['#1f77b4','#2ca02c'],
    wedgeprops={'edgecolor': 'white'},
    textprops={'fontsize': 12}
)

ax.legend(
    wedges,
    labels,
    title="Tipo de Membro",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

plt.title('Distribuição de Viagens por Tipo de Membro')
plt.axis('equal')  # Círculo perfeito
plt.tight_layout()
plt.show()
```

Dias da semana

```python
#Caminho do arquivo
cmse= '/content/drive/MyDrive/Analises/2022/dias_semana.csv'
dfse = pd.read_csv(cmse)
```

```python
#Caminho do arquivo
ordem_dias = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dfse['dia_semana'] = pd.Categorical(dfse['dia_semana'], categories=ordem_dias, ordered=True)
cores = {
    'member': '#2ca02c',  # verde
    'casual': '#1f77b4'   # azul
}


sns.lineplot(
    data=dfse,
    x='dia_semana',
    y='total',
    hue='tipo_uso',
    palette=cores,
    marker='o',
    linewidth=2,
    errorbar=None,
)

plt.title('Total de Viagens por Dia da Semana e Tipo de Membro')
plt.xlabel('Dia da Semana')
plt.ylabel('Total de Viagens')
plt.xticks(rotation=0)
plt.legend(title='Tipo de Membro')
```
